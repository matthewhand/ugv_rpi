# Gimbal Control, Robotic Arm Control, and LED Control
## Gimbal Control
### Basic Introduction

Products with a gimbal include two servos: a horizontal servo and a tilt servo. The horizontal servo controls the rotation of the gimbal in the horizontal plane, with a rotation range of ±180° (total range 360°); the tilt servo controls the rotation of the gimbal in the vertical plane, with a rotation range of -30° to 90° (total range 120°).

For products without a built-in gimbal, it is also possible to add a gimbal to the UGV by yourself.

```python
# Input the library for base control
from base_ctrl import BaseController

# Function for detecting Raspberry Pi
def is_raspberry_pi5():
    with open('/proc/cpuinfo', 'r') as file:
        for line in file:
            if 'Model' in line:
                if 'Raspberry Pi 5' in line:
                    return True
                else:
                    return False

# Determine the GPIO Serial Device Name Based on the Raspberry Pi Model
if is_raspberry_pi5():
    base = BaseController('/dev/ttyAMA0', 115200)
else:
    base = BaseController('/dev/serial0', 115200)
```

In the code block above, we import and instantiate the library used for chassis control. Next, we control the gimbal movement by changing the angles of the gimbal's horizontal and tilt servos.

Change the values in the following code. In this code:
- input_x is the rotation angle of the gimbal's horizontal servo. The horizontal servo PAN axis has a rotation range of ±180° (total range 360°);
- input_y is the rotation angle of the gimbal's tilt servo. The tilt servo TILT axis has a rotation range of -45° to 90° (total range 135°);
- input_speed is the speed at which the gimbal rotates. When the speed value is 0, the rotation is at its fastest;
- input_acc is the acceleration at the start and end of the gimbal's movement. The smaller the value, the smoother the start and stop. When the acceleration value is 0, the rotation is performed at the maximum acceleration.

Run the following code, and you will see the gimbal move 45° to the right and 45° upward before stopping.

```python
input_x = 45
input_y = 45
input_speed = 0
input_acc = 0

base.gimbal_ctrl(input_x, input_y, input_speed, input_acc)
```

In addition to controlling the gimbal's movement by changing the rotation angles of its two servos, you can also directly control the gimbal to move continuously.

Change the values in the following code. In this code:

- input_x is the horizontal servo rotation mode of the gimbal. When the value is -1, it means the gimbal continuously rotates to the left (clockwise); when the value is 1, it means the gimbal continuously rotates to the right (counterclockwise); when the value is 0, it means to stop rotating.
- input_y is the tilt servo rotation mode of the gimbal. When the value is -1, it means the gimbal's tilt servo continuously tilts downward (looking down); when the value is 1, it means the tilt servo continuously tilts upward (looking up); when the value is 0, it means to stop rotating.
- input_speed is the speed of the gimbal rotation.

The gimbal will automatically return to its middle position only when both input_x and input_y are set to 2.

Run the following code, and the gimbal will move to the left until it reaches 180° and then stop.

```python
input_x = -1
input_y = 0
input_speed = 0

base.gimbal_base_ctrl(input_x, input_y, input_speed)
```

Run the following code, and the gimbal will move upward and stop when it reaches 90°.

```python
input_x = 0
input_y = 1
input_speed = 0

base.gimbal_base_ctrl(input_x, input_y, input_speed)
```

Finally, run the following code to return the gimbal to the center position.

```python
input_x = 2
input_y = 2
input_speed = 0

base.gimbal_base_ctrl(input_x, input_y, input_speed)
```

## LED Control

### Basic Introduction 
The driver board integrates two 12V switching control interfaces (the actual maximum voltage will change with the battery voltage). These are controlled by ESP32's IO4 and IO5 pins via MOSFETs. Each channel has two interfaces, for a total of four 12V switching control interfaces. In the default assembly, IO4 controls the chassis headlights, and IO5 controls the pan-tilt headlights. You can control the switching of these two interfaces and adjust the voltage level by sending corresponding commands to the lower-level computer (however, due to the inherent delay in MOSFET control, the PWM output of the ESP32's IO pins is not linearly related to the actual voltage output of the interface).

For products without built-in LEDs, you can add 12.6V LEDs to the two 12V switch control interfaces (generally, 12V LEDs are also acceptable, as the UPS will not charge the battery above 12V for safety and battery protection). You can also add other peripherals to the remaining switch control interfaces—for example, a 12V water gun gearbox, which can be directly connected to the IO5 control interface to achieve automatic aiming and firing functions.

To run the program within a code block, select the code block and press Ctrl+Enter. If you are using JupyterLab on a mobile phone or tablet, you can press the triangle play button at the top to run the code block.

In the code block above, we imported and instantiated the library used to control the chassis. Next, we control the switch of the IO4 interface. The variable IO4_PWM is the PWM output of the ESP32's IO4 pin, with a range of 0-255. When this variable is 0, the switch controlled by IO4 is off; when this variable is 255, the voltage output by the switch controlled by IO4 is close to the BAT voltage of the UPS (the current voltage of the three series-connected lithium batteries inside the UPS).

Run the following code block to turn on the switch controlled by IO4 (turning on the chassis headlights). Note: The headlight is located on the camera gimbal. If the product is not equipped with a camera gimbal, there will be no headlight.

```python
IO4_PWM = 255
IO5_PWM = 0

base.lights_ctrl(IO4_PWM, IO5_PWM)
```

Run the following code block to turn on the IO5-controlled interface switch (which lights up the gimbal headlamp). 
Note: If the product is not equipped with a camera gimbal, there will be no headlamp.

```python
IO4_PWM = 255
IO5_PWM = 255

base.lights_ctrl(IO4_PWM, IO5_PWM)
```

If your product has LED lights, they should all be lit by now. You can run the following code block to dim the LED lights.

```python
IO4_PWM = 64
IO5_PWM = 64

base.lights_ctrl(IO4_PWM, IO5_PWM)
```

Finally, run the following code block to turn off the LED light.

```python
base.lights_ctrl(0, 0)
```

This runs a code block that integrates gimbal functionality

```python
import time
base.gimbal_ctrl(0, 0, 0, 0)
base.lights_ctrl(255, 255)
time.sleep(1)
base.gimbal_ctrl(45, 0, 0, 0)
base.lights_ctrl(0, 0)
time.sleep(1)
base.gimbal_ctrl(-45, 0, 0, 0)
base.lights_ctrl(255, 255)
time.sleep(1)
base.gimbal_ctrl(0, 90, 0, 0)
base.lights_ctrl(0, 0)
time.sleep(1)
base.gimbal_ctrl(0, 0, 0, 0)
```
