对于没有配有机械臂的产品，也可以自行在 UGV 上扩展机械臂。

```python
# 导入用于控制底盘的库
from base_ctrl import BaseController

# 用于检测树莓派的函数
def is_raspberry_pi5():
    with open('/proc/cpuinfo', 'r') as file:
        for line in file:
            if 'Model' in line:
                if 'Raspberry Pi 5' in line:
                    return True
                else:
                    return False

# 根据树莓派的型号来确定 GPIO 串口设备名称
if is_raspberry_pi5():
    base = BaseController('/dev/ttyAMA0', 115200)
else:
    base = BaseController('/dev/serial0', 115200)
```

```python
# 可以根据自己手中产品的类型，更改下面代码块中的JSON参数，来配置底盘类型：

# 底盘类型选择

# 其中“main”的值为底盘类型：
# - 1：RaspRover(四轮四驱底盘)
# - 2：UGV Rover(六轮四驱底盘)
# - 3：UGV Beast(履带底盘)  

# “module”的值为模块类型：
# - 0：没有安装模块和安装有云台模块都可以设置为该值
# - 1：M2机械臂模块（如果没有安装机械臂会导致底盘Ping不通关节报错）
# - 2：云台模块（如果没有安装云台会导致底盘Ping不同关节报错）
# - 3：M3机械臂模块（如果没有安装机械臂会导致底盘Ping不通关节报错）

#对于 m2 机械臂
base.base_json_ctrl({"T":900,"main":2,"module":1})

# #对于 m3 机械臂
# base.base_json_ctrl({"T":900,"main":2,"module":1})
```

在上面的代码块我们导入并实例化用于底盘控制的库，正确设置底盘和模块类型，接下来通过改变各个关节机的角度来控制机械臂运动。

改变以下代码中的值，以下代码中：
- base 为基础关节舵机的转动弧度，转动范围 ±180°（总范围360°）；
- shoulder 为肩关节舵机的转动弧度，转动范围 ±90°（总范围180°）；
- elbow 为肘关节舵机的转动弧度，转动范围 ±90°（总范围180°）；
- wrist 为肘关节舵机的转动弧度，转动范围 ±90°（总范围180°）；
- roll 为肘关节舵机的转动弧度，转动范围 ±180°（总范围360°）；
- hand 为末端关节舵机的转动弧度，转动范围 90°（总范围135°）。

运行以下代码，可以看见机械臂从当前位置运动到指定初始位置后停止。

```python
base_radian = 0.0
shoulder_radian = 0.0
elbow_radian = 1.5708
wrist_radian = 0.0
roll_radian = 0.0
hand_radian = 3.1415926
spd = 0.0
acc = 30.0    

# 对于 m2 机械臂
base.base_json_ctrl({"T":102,"base": base_radian,"shoulder": shoulder_radian,"elbow": elbow_radian,"hand": hand_radian,"spd": spd,"acc": acc})

# # 对于 m3 机械臂
# base.base_json_ctrl({"T":102,"base": base_radian,"shoulder": shoulder_radian,"elbow": elbow_radian,"wrist": wrist_radian,"roll": roll_radian,"hand": hand_radian,"spd": spd,"acc": acc})
```
