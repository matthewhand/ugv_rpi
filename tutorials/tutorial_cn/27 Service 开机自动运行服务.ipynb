

### 使用 systemd 配置开机自动运行服务

Linux 使用 **systemd** 来管理后台服务，包括启动、停止、重启以及让程序开机自启。
只需创建一个 `.service` 文件，即可让 Python 程序或脚本自动动。

## 1. 创建用户级 Service 文件

用户级服务文件存放在：

```
~/.config/systemd/user/
```

如果目录不存在，创建它：

```bash
mkdir -p ~/.config/systemd/user
```

创建服务文件（例如：`ugv-app.service`）：

```bash
nano ~/.config/systemd/user/ugv-app.service
```

写入以下内容：

```ini
[Unit]
Description=UGV Python App
After=sound.target pipewire.service
Wants=pipewire.service

[Service]
ExecStart=/home/ws/ugv_rpi/ugv-env/bin/python /home/ws/ugv_rpi/app.py
Restart=always
Environment=XDG_RUNTIME_DIR=/run/user/%U
WorkingDirectory=/home/ws/ugv

[Install]
WantedBy=default.target
```

## 2. 重新加载 sys配置

```bash
systemctl --user daemon-reload
```

## 3. 允许开机自启

```bash
systemctl --user enable ugv-app.service
```

## 4. 手动启动服务

```bash
systemctl --user start ugv-app.service
```

## 5. 查看服务运行状态

```bash
systemctl --user status ugv-app.service
```

## 6. 查看运行日志

```bash
journalctl --user -u ugv-app.service -f
```
